```python
# Q5/ Housing prices prediction
```

```python
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data",
                   delim_whitespace = True, header=None,
                   names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM',
                            'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'])
```

```python
###     CRIM      per capita crime rate by town
#    2. ZN        proportion of residential land zoned for lots over 
#                 25,000 sq.ft.
#    3. INDUS     proportion of non-retail business acres per town
#    4. CHAS      Charles River dummy variable (= 1 if tract bounds 
#                 river; 0 otherwise)
#    5. NOX       nitric oxides concentration (parts per 10 million)
#    6. RM        average number of rooms per dwelling
#    7. AGE       proportion of owner-occupied units built prior to 1940
#    8. DIS       weighted distances to five Boston employment centres
#    9. RAD       index of accessibility to radial highways
#    10. TAX      full-value property-tax rate per $10,000
#    11. PTRATIO  pupil-teacher ratio by town
#    12. B        1000(Bk - 0.63)^2 where Bk is the proportion of blacks 
#                 by town
#    13. LSTAT    % lower status of the population
#    14. MEDV
```

```python
data
```

```python
indep_vars = ['LSTAT']
dep_vars = ['MEDV']
indep_data = data[indep_vars]
dep_data = data[dep_vars]
indep_train, indep_test, dep_train, dep_test = train_test_split(indep_data, dep_data, test_size=0.33, random_state=42)
```

```python
regr = linear_model.LinearRegression()
regr.fit(indep_train, dep_train)
```

```python
#print('Coefficients: {0}'.format(zip(indep_vars, np.squeeze(regr.coef_))))
np.squeeze(regr.coef_)
```

```python
# the coefficient of lstat is -0.9805 which means lstat affects medv negatively as lstat increases medv decreases
```

```python
sns.set(color_codes=True)
ax = sns.regplot(x="LSTAT", y="MEDV", data=data)
```

```python
# as the negative coefficient shows we can see from the scatter plot above that MEDV decreases as LSTAT increases
```
